Problem Statement:

The average IMDb rating of American tv shows of this year hasn't increased significatly with a significance level of **0.05**

H0: Sample mean (mean of tvshows released in 2022) = Population mean

H1: Sample mean > population mean

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('tv_shows.csv')
df
```

```python
IMDb=df[['Year','IMDb']]
IMDb
```

```python
IMDb.isnull().sum()
```

```python
IMDb.dropna(inplace=True)
IMDb
```

```python
IMDb.isnull().sum()
```

```python
IMDb_sample=IMDb[IMDb['Year']=='2020']
IMDb_sample
```

```python
def change(x):
    if x=='16+':
        x=float(0)
    else:
        x=str(x)[0:3]
        x=float(x)
    return x
```

```python
IMDb['IMDb']=IMDb['IMDb'].apply(change)
IMDb
```

```python
IMDb_sample['IMDb']=IMDb_sample['IMDb'].apply(change)
IMDb_sample
```

```python
import scipy.stats as st
```

```python
n=445
xbar=IMDb_sample['IMDb'].mean()
mu=IMDb['IMDb'].mean()
alpha=0.05
sigma=IMDb['IMDb'].std()
```

```python
z_critical = abs(st.norm.ppf(alpha/2))
z_critical
```

```python
z=(xbar-mu)/(sigma/np.sqrt(n))
z
```

```python
if z<z_critical and z>(-1*z_critical):
    print("Null hypothesis cannot be Rejected")
else:
    print("Null hypothesis is rejected")
```
